# get the proportion of stromal cells in scrna dataset

```r
scrna = readr::read_rds('../Cleaned scRNA references for cell typing/complete_pelka_dataset.rds')
scrna
```

```r
head(scrna)
```

```r
total_cells = nrow(scrna@meta.data)
total_cells
```

```r
stromal_cells = scrna@meta.data %>% filter(clTopLevel == 'Strom') %>% nrow()
stromal_cells
```

```r
percent_stromal_scrna = 100*stromal_cells/total_cells
percent_stromal_scrna
```

```r
unique(scrna@meta.data$clTopLevel)
```

```r
immune_cell_types = c('TNKILC', 'Myeloid', 'Mast', 'Plasma', 'Bplasma')
```

```r
immune_scrna = scrna@meta.data %>% filter(clTopLevel %in% immune_cell_types) %>% nrow()
immune_scrna
```

```r
percent_immune_scrna = 100*immune_scrna/total_cells
percent_immune_scrna
```

```r
merfish_metadata = readr::read_rds('/n/data1/bwh/medicine/korsunsky/lab/mup728/CRC_MERFISH_niches/labeled_seurat_objects/renamed_cell_states/metadata_complete.rds')
```

```r
glimpse(merfish_metadata)
unique(merfish_metadata$knn_coarse) 
```

```r
total_cells_merfish = nrow(merfish_metadata)
total_cells_merfish
```

```r
stromal_cells_merfish = merfish_metadata %>% filter(knn_coarse == 'Strom') %>% nrow()
stromal_cells_merfish
```

```r
immune_merfish = merfish_metadata %>% filter(knn_coarse %in% immune_cell_types) %>% nrow()
immune_merfish
```

```r
percent_strom_merfish = 100*stromal_cells_merfish/total_cells_merfish
percent_strom_merfish
```

```r
percent_immune_merfish = 100*immune_merfish/total_cells_merfish
percent_immune_merfish
```

s expected, the MERFISH dataset had a higher stromal cell frequency (24%) compared to the scRNA-seq reference (4.2%), which underrepresented them due to the mild tissue dissociation protocol. Conversely, the scRNA-seq data was enriched for immune cells (43.4% vs. 11.3% in MERFISH) via CD45+ magnetic sorting (Fig. 1C).
