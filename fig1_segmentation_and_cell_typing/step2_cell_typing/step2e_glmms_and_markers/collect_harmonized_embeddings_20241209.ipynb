# collect harmonized embeddings

```r
require(Seurat)
require(tidyverse)
require(glue)
```

```r
list.dirs(recursive = FALSE)
```

```r
fine_types = c('B', 'Epi', 'Myeloid', 'Plasma', 'Strom', 'TNKILC')
```

```r
fine_type = 'B'
i = 1
temp = readr::read_rds(seurat_files[i])
temp = Embeddings(temp, reduction = 'harmony') %>%
    as.data.frame() %>%
    mutate(orig.ident = temp@meta.data[rownames(.), 'orig.ident'], 
           technology = temp@meta.data[rownames(.), 'technology'], 
           combined_cell_names = temp@meta.data[rownames(.), 'combined_cell_names'], .before = 'harmony_1') %>%
    cbind(Embeddings(temp, reduction = 'humap')[rownames(.), ]) 
colnames(temp) %>% writeLines
```

```r
for (fine_type in fine_types){
    message(fine_type)
    seurat_files = list.files(path = fine_type, recursive = TRUE, full.names = TRUE)
    seurat_files = seurat_files[grepl(pattern = '_fine_types.*rds', x = seurat_files)]
    for (i in 1:length(seurat_files)){
        message(seurat_files[i])
        temp = readr::read_rds(seurat_files[i])
        Embeddings(temp, reduction = 'harmony') %>%
            as.data.frame() %>%
            mutate(orig.ident = temp@meta.data[rownames(.), 'orig.ident'], 
                   technology = temp@meta.data[rownames(.), 'technology'], 
                   combined_cell_names = temp@meta.data[rownames(.), 'combined_cell_names'], .before = 'harmony_1') %>%
            cbind(Embeddings(temp, reduction = 'humap')[rownames(.), ]) %>%
            data.table::fwrite(x = ., file = seurat_files[i] %>% gsub(pattern = '.rds', replacement = '_embeddings.csv'))    
    }
}
```

```r
coarseTypedSeurats = list.files(COARSE_TYPED_MSI_DIR, full.names = TRUE)[grep(x = list.files(COARSE_TYPED_MSI_DIR, full.names = TRUE), pattern = "coarse.*.rds")]
```

```r
for (i in 1:length(coarseTypedSeurats)){
    message(coarseTypedSeurats[i])
    temp = readr::read_rds(coarseTypedSeurats[i])
    Embeddings(temp, reduction = 'harmony') %>%
        as.data.frame() %>%
        mutate(orig.ident = temp@meta.data[rownames(.), 'orig.ident'], 
               technology = temp@meta.data[rownames(.), 'technology'], 
               combined_cell_names = temp@meta.data[rownames(.), 'combined_cell_names'], .before = 'harmony_1') %>%
        cbind(Embeddings(temp, reduction = 'humap')[rownames(.), ]) %>%
        data.table::fwrite(x = ., file = coarseTypedSeurats[i] %>% gsub(pattern = '.rds', replacement = '_embeddings.csv'))
    rm(temp);gc()
}
```
