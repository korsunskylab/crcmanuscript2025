```r
getCoarseTypedObject = function(temp, name){
    temp = readr::read_rds(temp)
    temp = subset(temp, technology == 'MERFISH')
    temp@meta.data = temp@meta.data[, c('orig.ident', 'x', 'y', 'cell_id', 'ClusterTop', 'condition')]
    write_rds(x = temp, glue::glue('/n/data1/bwh/medicine/korsunsky/lab/mup728/CRC_MERFISH_niches/coarse_', name,'.rds'))
}
```

```r
sampleIDs = gsub(x = coarseTypedSeurats, pattern = '/n/scratch/users/m/mup728/mup728/Pelka_Baysor_segmentation/coarse_typing_with_weighted_knn/Coarse_typing_with_weighted_knn/MSI/coarse_|.rds', replacement = '')
sampleIDs
```
